```bash
# nmap y zmap están disponibles en los repositorios oficiales de Ubuntu 24.04 LTS
sudo apt update && sudo apt install -y nmap zmap
```

```bash
# masscan no está en los repositorios de Ubuntu, por lo que se compila desde el código fuente
sudo apt-get install -y git make gcc
```

```bash
git clone https://github.com/robertdavidgraham/masscan
cd masscan && make
```

```bash
# Escaneo de todos los puertos TCP con nmap
# -p 1-65535 cubre el rango completo de puertos posibles
nmap -p 1-65535 104.16.231.132
```

```bash
# Escaneo equivalente con masscan, que opera a más bajo nivel y requiere permisos de root
# --rate 100 limita la tasa de envío para evitar saturar la interfaz de red de la VM
sudo masscan -p 1-65535 --rate 100 104.16.231.132
```

```bash
# zmap está diseñada para escanear un puerto concreto sobre un rango amplio de direcciones IP,
# no para enumerar puertos de un único host, por lo que no resulta adecuada para este laboratorio.
# Uso típico de referencia:
#   sudo zmap -p 443 192.168.1.0/24
```

```bash
# Escaneo en profundidad sobre el puerto 80
# -sV identifica el software que atiende el puerto, -O intenta determinar el sistema operativo
sudo nmap -p 80 -sV -O 104.16.231.132
```
